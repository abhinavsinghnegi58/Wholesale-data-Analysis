### Handling null values in  numeric columns

* Now after finding null Values in each columns to fix those null values and missing values we have to check for Outliers

1. If outliers are present then replace the null values by median of the columns 
2. If ouutliers are not present then replace the null values by the median of the column
3. For catagorical column repllace the null/missing values by mode of the column

* checking for outliers

```python
list=['Fresh',"Delicatessen","Frozen",'Detergents_Paper','Milk','Grocery']
for i in list:
    plt.figure(figsize=(10,3))
    df.boxplot(column=i)
    plt.show()
```

We can see there are outliers in all the numerical columns

* replacing nan values with median, because outliers are present in columns
* after replacing nan values to their median we have done our outlier treatment for numerical columns

```python
list=['Milk','Grocery','Frozen','Detergents_Paper','Fresh','Delicatessen']
```

```python
def rep_nan(l):
    for i in l:
        med=df[i].median()
        df[i].replace(np.nan,med,inplace=True)
```

```python
rep_nan(list)
```

```python
def remove_outliers(columns):
        sorted(columns)
        Q1=columns.quantile(.25)
        Q3=columns.quantile(.75)
        IQR=Q3-Q1
        lb=Q1-(1.5*IQR)
        ub=Q3+(1.5*IQR)
        return lb,ub
```

```python
flb,fub=remove_outliers(df["Fresh"])
df["Fresh"]=np.where(df["Fresh"]<flb,flb,df["Fresh"])
df["Fresh"]=np.where(df["Fresh"]>fub,fub,df["Fresh"])
```

```python
dlb,dub=remove_outliers(df["Delicatessen"])
df["Delicatessen"]=np.where(df["Delicatessen"]<dlb,dlb,df["Delicatessen"])
df["Delicatessen"]=np.where(df["Delicatessen"]>dub,dub,df["Delicatessen"])
```

```python
fzlb,fzub=remove_outliers(df["Frozen"])
df["Frozen"]=np.where(df["Frozen"]<fzlb,fzlb,df["Frozen"])
df["Frozen"]=np.where(df["Frozen"]>fzub,fzub,df["Frozen"])
```

```python
dplb,dpub=remove_outliers(df["Detergents_Paper"])
df["Detergents_Paper"]=np.where(df["Detergents_Paper"]<dplb,dplb,df["Detergents_Paper"])
df["Detergents_Paper"]=np.where(df["Detergents_Paper"]>dpub,dpub,df["Detergents_Paper"])
```

```python
flb,fub=remove_outliers(df["Grocery"])
df["Grocery"]=np.where(df["Grocery"]<flb,flb,df["Grocery"])
df["Grocery"]=np.where(df["Grocery"]>fub,fub,df["Grocery"])
```

```python
flb,fub=remove_outliers(df["Milk"])
df["Milk"]=np.where(df["Milk"]<flb,flb,df["Milk"])
df["Milk"]=np.where(df["Milk"]>fub,fub,df["Milk"])
```

# Buyers Insights

1. What is the repeat purchase rate for buyers who spend above the average in at least three categories?

* Repeat Purchase Rate for buyers who spend above the average in at least three categories: 41.82%

```python
categories = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']
average_spending = df[categories].mean()

# Identify buyers who spend above the average in at least three categories
above_average_spenders = df.apply(lambda row: sum(row[cat] > average_spending[cat] for cat in categories), axis=1)
repeat_buyers = df[above_average_spenders >= 3]

# Calculate the repeat purchase rate
# Assuming repeat purchase is indicated by multiple entries in the dataset
repeat_purchase_rate = len(repeat_buyers) / len(df) * 100
```

```python
print(f"Repeat Purchase Rate for buyers who spend above the average in at least three categories: {repeat_purchase_rate:.2f}%")
```
